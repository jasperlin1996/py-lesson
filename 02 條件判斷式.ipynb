### if-else
* We can use if-else statements to define our logical judgment in our program.
* 每個 `else` 都都需要有一個 `if` 或是 `elif` 在他之前



#### Usage
    if(判斷式):
        blablabla...
    elif(判斷式):
        blablabla...
    else:
        blablabla...
        
* 如果你希望在 `if-else` 的執行區塊內不要做式的話記得要用 `pass`

```python
if(a):
    print("{} is True".format(a))
```

```python
if('Z'):
    print("{} is True".format('Z'))
```

```python
if(()):
    print("{} is True".format('()'))
else:
    print("{} is not True(False)".format('()'))
```

```python
if (num % 4 == 0):
    print("{} is divisible by 4".format(num))
elif (num % 3 == 0):
    print("{} is divisible by 3 but not 4".format(num))
else:
    print("{} isn't divisible by both 3 or 4".format(num))
```

### HW2 平年還是閏年(Common or Leap year)

>  We all know that usually there are 365 days a year. But when it comes to a "Leap year", it has 366 days instead of 365, by extending February to 29 days rater than the common 28.

> Write a python program that helps us find out if the input year is a leap year or a common year.

##### Tips
1. if (year is not divisible by 4) then (it is a common year)
2. else if (year is not divisible by 100) then (it is a leap year)
3. else if (year is not divisible by 400) then (it is a common year)
4. else (it is a leap year)
