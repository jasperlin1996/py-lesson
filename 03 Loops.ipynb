## Loops
* 迴圈可以允許我們執行一段程式碼多次
* 下圖為基本的迴圈邏輯結構

* Python 中的迴圈有兩種
    
        while
        for
    
* 並且在迴圈中 Python 還有提供了控制語句

        break
        continue
        pass

### While loop
    while(條件式):
        執行語句


```python
count = 0
while(count < 10):
    print("count = {}".format(count))
    count += 1
```

### For loop
* Python 的 for 迴圈可以遍歷任何序列的物件，如一個 list 或者一個 str 等等。
* 實例如下

    for iterating_var in sequence:
        statements(s)
